```python
paths = glob.glob(r'C:\Users\HUAWEI\Desktop\Final_project\T5\CUSTOM_DATASET\*\*')

def read_folders(paths):
    audios = []
    
    for item_path in paths:
        if item_path.endswith('.wav'):  
            print('audio:', item_path)
            audio, sr = librosa.load(item_path)
            audios.append(audio)
            print(audio, "------", sr)
            print(audios)
    return audios
```
