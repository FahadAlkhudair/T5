```python
data = '/content/drive/MyDrive/Final_Project/T5/sounds_final'


dir_list = os.listdir(data)
dir_list.sort()

classes = []
path = []
for i in dir_list:

    sub_dir_path = os.path.join(data, i)

    if os.path.isdir(sub_dir_path):
        fname = os.listdir(sub_dir_path)
        for f in fname:

            if f.endswith('.wav'):
                    classes.append(i)
                    file_path = os.path.join(sub_dir_path, f)
                    path.append(file_path)
```

```python
df = pd.DataFrame({'classes': classes, 'path': path})

df
```

```python
import whisper
import pandas as pd


model = whisper.load_model("large")

```

```python
def transcribe_audio(audio_path):
    result = model.transcribe(audio_path)
    return result['text']


df['transcription'] = df['path'].apply(transcribe_audio)
df
```
