```python
data = '/content/drive/MyDrive/T5/sounds'


dir_list = os.listdir(data)
dir_list.sort()

classes = []
path = []
for i in dir_list:

    sub_dir_path = os.path.join(data, i)

    if os.path.isdir(sub_dir_path):
        fname = os.listdir(sub_dir_path)
        for f in fname:

            if f.endswith('.wav'):
                    classes.append(i)
                    file_path = os.path.join(sub_dir_path, f)
                    path.append(file_path)
```

```python
df = pd.DataFrame({'classes': classes, 'path': path})

df
```

```python
import whisper
import pandas as pd


model = whisper.load_model("large-v3")

```

```python
def transcribe_audio(audio_path):
    result = model.transcribe(audio_path)
    return result['text']


df['transcription'] = df['path'].apply(transcribe_audio)
df
```

```python
import pandas as pd
from transformers import pipeline


# Initialize the zero-shot classification pipeline
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Function to apply classification
def classify_text(row):
    # Extract classes as a list of strings, assuming they are separated by commas in the DataFrame
    classes = row['classes'].split(',')  # Adjust this if your format is different
    # Apply the classifier
    result = classifier(row['transcription'], classes)
    # Return whatever part of the result you're interested in, e.g., labels and scores
    return result['labels'][0], result['scores'][0]  # Example: return top label and its score

# Apply the classification function to each row
# Note: This can be time-consuming for a large number of rows
df['classification_result'] = df.apply(classify_text, axis=1)

# Optionally, split the results into separate columns if needed
df[['top_label', 'top_score']] = pd.DataFrame(df['classification_result'].tolist(), index=df.index)
```
